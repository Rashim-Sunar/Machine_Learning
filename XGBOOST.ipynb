# What is XGBoost?
#### XGBoost (Extreme Gradient Boosting) is an advanced ensemble learning algorithm based on gradient boosting.

#### It builds models:
#### Sequentially, where each new model corrects the errors of previous models.
#### XGBoost improves traditional gradient boosting by making it:
<ul>
    <li>Faster</li>
    <li>More accurate</li>
    <li>Regularized</li>
    <li>Highly optimized</li>
</ul>

# Core Idea of XGBoost
#### XGBoost works by:
<ol type='1'>
    <li>Building trees sequentially</li>
    <li>Minimizing a loss function</li>
    <li>Using gradient descent</li>
    <li>Adding new trees to correct previous errors</li>
</ul>

#### Final model:
``` F(x)= ∑f​(x), where k=1 to K```
#### where, f(x) = decision tree
#### k = no. of trees
#### Each new tree learns from previous mistakes.

---

# How XGBoost Works (Step-by-Step)
### Step 1: Initialize Model
#### Start with a simple prediction.
#### Example:
<ul>
    <li>Mean value (regression)</li>
    <li>Log odds (classification)</li>
</ul>

### Step 2: Compute Residuals
#### Calculate errors between:
``` Actual value – Predicted value ```
#### These errors are called residuals.

### Step 3: Train First Tree
#### Train decision tree to predict residuals.
#### The tree learns: How much prediction should be adjusted.

### Step 4: Update Predictions
#### Update model prediction:
``` y^​ = y^ + learning_rate × tree_prediction ```
#### Learning rate controls step size.

### Step 5: Repeat
#### Repeat process:
<ul>
    <li>Calculate new residuals</li>
    <li>Train next tree</li>
    <li>Update predictions</li>
</ul>

#### Until:
<ul>
    <li>Maximum trees reached</li>
    <li>Error minimized</li>
</ul>

## What Makes XGBoost Special?
#### XGBoost improves Gradient Boosting with:

### 1. Regularization

### 2. Tree Pruning
#### Removes splits that do not improve model performance.
#### Uses gain threshold.

### 3. Parallel Processing
#### Unlike traditional boosting, XGBoost can parallelize tree construction.
#### Much faster.

### 4. Handling Missing Values
#### Automatically learns best direction for missing values.

### 5. Built-in Cross Validation
#### XGBoost has internal CV functions.

# Important Hyperparameters

## 1. n_estimators
#### Number of trees.
#### More trees → stronger model.

## 2. learning_rate (eta)
#### Controls contribution of each tree.
#### Small learning rate → slower but better generalization.
#### Typical: 
```0.01 – 0.3 ```

## 3. max_depth
#### Maximum depth of tree.
#### Controls complexity.

## 4. subsample
#### Percentage of training samples used per tree.
#### Helps prevent overfitting.

## 5. colsample_bytree
#### Percentage of features used per tree.
#### Adds randomness.

## 6. gamma
#### Minimum loss reduction required for split.
#### Controls pruning.
