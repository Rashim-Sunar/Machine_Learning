# Bias
#### The inability of machine learning model to truely capture the relationship in training data.

#### Bias = Error due to wrong assumptions.
#### It measures:
#### How far your models predictions are from the true underlying relationship.

---

# Variance
#### Variance = Sensitivity to training data.

#### It measures:
#### How much the model changes if you train it on different datasets.

---

## BiasVariance Trade-off
#### BiasVariance Trade-off explains:
#### Why a model underfits or overfits and how to balance model complexity.

## Core Idea
### Total prediction error can be decomposed into:
#### We want:
<ul>
    <li> Low Bias</li>
    <li> Low Variance</li>
</ul>

#### But reducing one often increases the other.
#### Thats the trade-off.

---

## How to Reduce High Bias?
####  Increase model complexity 
####  Add features
####  Use polynomial features
####  Reduce regularization

---

## How to Reduce High Variance?
####  Regularization (L1, L2)
####  Bagging (Random Forest)
####  Boosting
####  More training data
####  Reduce model complexity
