# Ensemble Learning
#### Ensemble Learning is a machine learning technique where:
#### Multiple models (weak learners) are combined to create a stronger model.

#### Instead of relying on one model, we combine several models to improve:
<ul>
    <li>Accuracy</li>
    <li>Stability</li>
    <li>Generalization</li>
</ul>

---

## Why Do We Use Ensemble Learning?
#### Single models may suffer from:
<ul>
    <li>High variance (overfitting)</li>
    <li>High bias (underfitting)</li>
    <li>Sensitivity to noise</li>
</ul>

#### Ensemble methods help reduce:
<ul>
    <li>Variance</li>
    <li>Bias</li>
    <li>Or both</li>
</ul>

---

## Types of Ensemble Methods
#### There are three major types:
<ol>
    <li>Bagging</li>
    <li>Boosting</li>
    <li>Stacking</li>
</ol>

---

# 1. Bagging (Bootstrap Aggregating)
#### Train multiple models in parallel on different random samples of data.
#### Final prediction:
<ul>
    <li>Average (for regression)</li>
    <li>Majority vote (for classification)</li>
</ul>

## Example: Random Forest
#### Random Forest:
<ul>
    <li>Many decision trees</li>
    <li>Each trained on random subset</li>
    <li>Predictions averaged</li>
</ul>

#### Reduces: Variance
####  Does NOT reduce: Bias much

---

# 2. Boosting
#### Train models sequentially.
#### Each new model: Focuses on correcting previous model’s errors.

### Popular Boosting Algorithms
<ul>
    <li>AdaBoost</li>
    <li>Gradient Boosting</li>
    <li>XGBoost</li>
    <li>LightGBM</li>
    <li>CatBoost</li>
</ul>

#### Reduces: Bias
#### May increase: Variance (if overfitted)

---

# 3. Stacking
#### Train multiple different models
#### Combine their outputs using a meta-model
#### Example:
<ul>
    <li>Model 1 → Logistic Regression</li>
    <li>Model 2 → Random Forest</li>
    <li>Model 3 → SVM</li>
    <li>Meta-model → learns from their predictions</li>
</ul>
