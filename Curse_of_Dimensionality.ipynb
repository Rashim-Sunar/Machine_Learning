# Curse of Dimensionality

## What is Curse of Dimensionality?
#### When the number of features (dimensions) increases:
<ul>
    <li>Data becomes sparse</li>
    <li>Model performance decreases</li>
    <li>Computation increases</li>
</ul>

#### In simple words:
Too many features can hurt the model.

### Why It Happens?
#### Imagine:
1D space → easy to find neighbors <br/>
2D space → slightly harder <br/>
100D space → everything becomes far apart <br/>

#### Distance between points increases drastically.

### This breaks:
<ul>
    <li>KNN</li>
    <li>Clustering</li>
    <li>Distance-based models</li>
</ul>

---

## Problems Caused
<ul>
    <li>Performance decreases</li>
    <li>Overfitting increases</li>
    <li>Computation cost increases</li>
    <li>Data sparsity</li>
</ul>

### Example

#### If each feature has 10 possible values:
#### With 2 features → 100 combinations
#### With 10 features → 10¹⁰ combinations

#### Data becomes extremely sparse.

---

## How to Handle It?
### 1️⃣ Feature Selection
#### Remove irrelevant features.

#### Methods:
<ul>
    <li>Forward Selection</li>
    <li>Backward Elimination</li>
    <li>Lasso</li>
</ul>


### 2️⃣ Feature Extraction
#### Transform features into fewer dimensions.

#### Methods:
<ul>
    <li>PCA</li>
    <li>LDA</li>
    <li>t-SNE</li>
</ul>


### 3️⃣ Regularization
<ul>
    <li>L1 (Lasso)</li>
    <li>L2 (Ridge)</li>
</ul>

---

## Algorithms Sensitive to It
<ul>
    <li>KNN (very sensitive)</li>
    <li>SVM</li>
    <li>Neural Networks</li>
</ul>

#### Tree models handle it better.
