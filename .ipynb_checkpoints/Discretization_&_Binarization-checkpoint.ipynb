## 2ï¸âƒ£ Equal Frequency Binning

## 3ï¸âƒ£ K-Means Binning

In sklearn:
```sh
from sklearn.preprocessing import KBinsDiscretizer

kbd = KBinsDiscretizer(
    n_bins=4,
    encode='ordinal',
    strategy='uniform'   # or 'quantile' or 'kmeans'
)
```
