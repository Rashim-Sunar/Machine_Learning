# AdaBoost (Adaptive Boosting)
#### AdaBoost (Adaptive Boosting) is an ensemble technique that: Combines multiple weak learners sequentially, where each new learner focuses more on previously misclassified samples.

---

## Core Idea of AdaBoost
#### Instead of training models independently (like Bagging):
#### AdaBoost:
<ol>
    <li>Trains models one after another.</li>
    <li>Adjusts weights of training samples.</li>
    <li>Gives more importance to wrongly classified samples.</li>
    <li>Combines models using weighted voting.</li>
</ol>

---

## What is a Weak Learner?
#### Usually:
<ul>
    <li>Decision Stump (tree with depth = 1)</li>
</ul>

#### Weak learner: Slightly better than random guessing.
#### AdaBoost converts many weak learners into a strong learner.

## Hyperparameters of AdaBoost
### n_estimators
Number of weak learners.
### learning_rate: 
Shrinks contribution of each learner.
Smaller learning rate:
<ul>
    <li>More stable</li>
    <li>Needs more estimators</li>
</ul>
