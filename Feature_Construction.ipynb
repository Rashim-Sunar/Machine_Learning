# 1️⃣ Feature Construction
## What is Feature Construction?
#### Feature Construction is the process of creating new features from existing features to improve model performance.
#### It is part of Feature Engineering.
#### Instead of using raw features directly, we build more informative ones.

### Example:
```sh
Length	  Width
  10	   5
```
#### Better feature:
```sh
Area = Length × Width
```
Area may be more useful than length and width separately.

---

##  Types of Feature Construction
### 1️⃣ Mathematical Transformation
<ul>
    <li>Log(x)</li>
    <li>x²</li>
    <li>√x</li>
    <li>Interaction terms (x1 * x2)</li>
</ul>

#### Example:
```sh
df['income_log'] = np.log(df['income'])
df['area'] = df['length'] * df['width']
```

### 2️⃣ Combining Features
#### Example:
```sh
Total_income = salary + business_income
```

### 3️⃣ Time-Based Construction

From:
```sh
Date: 2026-01-31
```

#### Create:
<ul>
    <li>Year</li>
    <li>Month</li>
    <li>Day</li>
    <li>Weekday</li>
    <li>Is_weekend</li>
</ul>

### 4️⃣ Domain-Based Construction
#### Example (Titanic):
```sh
Family_size = SibSp + Parch + 1
```

---

##  2️⃣ Feature Splitting
### What is Feature Splitting?
#### Feature Splitting means dividing one feature into multiple features.
#### Opposite of feature construction.

### Why Do We Use It?
#### Sometimes one column contains multiple pieces of information.

### Example:
```sh
Full_Name = "Rashim Sunar"
```

#### Split into:
<ul>
    <li>First_Name</li>
    <li>Last_Name</li>
</ul>

---

### Examples
#### 1️⃣ Date Splitting
```sh
2026-01-31
```
#### →
<ul>
    <li>Year</li>
    <li>Month</li>
    <li>Day</li>
</ul>

#### 2️⃣ Text Feature Splitting
```sh
Product_Code = ABC-123-XYZ
```
Split into:
<ul>
    <li>Category</li>
    <li>Number</li>
    <li>Region</li>
</ul>

#### 3️⃣ Address Splitting
```sh
Kathmandu, Nepal
```

##### Split into:
<ul>
    <li>City</li>
    <li>Country</li>
</ul>

#### Python Example
```sh
df[['city','country']] = df['location'].str.split(',', expand=True)
```

### Why Important?
<ul>
    <li>Because models cannot understand combined text directly.</li>
    <li>Splitting gives more structured information.</li>
</ul>
